Task 1

```python
import pandas as pd
# Load dataset
df = pd.read_csv("/content/sample_data/used_cars.csv")
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nDescribe:")
print(df.describe())
```

Task 2

```python
# 1. Drop rows where target is missing
df = df.dropna(subset=["selling_price"])
# 2. Handle missing values (example: fill with median)
df["mileage"] = df["mileage"].fillna(df["mileage"].mode()[0])
df["engine"] = df["engine"].fillna(df["engine"].median())
# 3. Clean 'brand' column
df["brand"] = df["brand"].str.strip().str.lower()
# 4. Extract numeric values from mileage
df["mileage"] = df["mileage"].astype(str).str.extract('(\d+\.?\d*)')
df["mileage"] = df["mileage"].astype(float)
# 5. Remove duplicates
df = df.drop_duplicates()
# Final check
print("Cleaned Shape:", df.shape)
print(df.head())
```

Task 3

```python
from sklearn.metrics import mean_absolute_error
y = df["selling_price"]
baseline_pred = [y.mean()] * len(y)
mae = mean_absolute_error(y, baseline_pred)
print("Baseline MAE:", mae)
```
